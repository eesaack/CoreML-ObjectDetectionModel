```python
import pandas as pd
data = pd.read_csv('ar41_data.csv')
```

```python
# Specify the number of rows you want in the sample
sample_size = 100000  # You can adjust this number

# Take a random sample using pandas
sample_df = data.sample(n=sample_size, random_state=42)  #random_state for reproducibility
```

```python
sample_df.to_csv('data-mining.csv', index=False)
```
