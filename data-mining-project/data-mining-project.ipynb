```python
df['timestamps_UTC'] = pd.to_datetime(df['timestamps_UTC'], utc=True)
# Convert UTC timestamp to belgian timestamp
dst_start = pd.Timestamp('2023-03-26 01:00:00', tz='UTC')
after_dst_start = df['timestamps_UTC'] >= dst_start
# For timestamps before DST start, add 1 hour, and for those after, add 2 hours
df['timestamps_Belgian'] = df['timestamps_UTC'] + pd.to_timedelta(after_dst_start * 1, unit='h')
del after_dst_start
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_timestamps(df):
  plt.figure(figsize=(10, 6))
  sns.histplot(df, bins=50, kde=True)
  plt.xlabel('Timestamps (Belgian Time)')
  plt.ylabel('Frequency')
  plt.title('Distribution of Timestamps')
  plt.show()

before_2023 = df[df['timestamps_Belgian'].dt.year < 2023]
print('Amount of data from 2022:', before_2023.shape[0], 'rows')
print('Percentage of data from 2022:', round(before_2023.shape[0]/df.shape[0] * 100, 5) ,'%')

# Removing the data from 2022
df = df[df['timestamps_UTC'].dt.year >= 2023]
plot_timestamps(df['timestamps_Belgian'])
```

```python
import folium

smaller_sample = df.copy().sample(5000)

def create_map(df):
  # Starting point for the map
  map_center_lat = df['lat'].mean()
  map_center_lon = df['lon'].mean()
  # Create map and add points
  train_map = folium.Map(location=[map_center_lat, map_center_lon], zoom_start=6)
  for idx, row in df.iterrows():
      folium.CircleMarker(location=[row['lat'], row['lon']],
                          radius=3,
                          color='purple',
                          fill=True,
                          fill_color='purple',
                          fill_opacity=0.6
                        ).add_to(train_map)

  return train_map
#train_map

create_map(smaller_sample)
```
