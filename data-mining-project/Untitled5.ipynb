```python
import pandas as pd
df = pd.read_csv('merged_temp_elevation.csv')
df.head()
```

```python
df['timestamps_UTC'] = pd.to_datetime(df['timestamps_UTC'])
df['date_hour'] = df['timestamps_UTC'].dt.to_period('H')
```

```python
df.groupby(['mapped_veh_id', 'date_hour']).agg({'lat': 'mean', 'lon': 'mean'}).reset_index()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
df.to_csv('tableau.csv', index=False)
```

```python
df.to_excel('tableau.xlsx', index=False)
```
