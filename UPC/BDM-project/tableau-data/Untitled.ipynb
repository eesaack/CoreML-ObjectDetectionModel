```python
import pandas as pd
import numpy as np
```

```python
def insert_age_column(file_path, output_file_path):
    data = pd.read_csv(file_path)
    n = len(data)
    age_brackets = [(18, 25), (26, 35), (36, 45)]
    probabilities = [0.523, 0.314, 0.163]
    np.random.seed(42)
    ages = np.random.choice(
        [np.random.randint(low, high + 1) for low, high in age_brackets],
        size=n,
        p=probabilities
    )
    
    age_column_index = data.columns.get_loc("name") + 1
    data.insert(loc=age_column_index, column='age', value=ages)
    data.to_csv(output_file_path, index=False)
    
file_path = 'user-data.csv'  
output_file_path = 'user-data.csv'  
insert_age_column(file_path, output_file_path)
```
